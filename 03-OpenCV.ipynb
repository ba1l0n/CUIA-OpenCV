```python
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import requests
```

```python
def plot(image):
    if len(image.shape)==2:
        h, w = image.shape
        c = 1
    else:
        h, w, c = image.shape
    px = 1/plt.rcParams['figure.dpi']  # pixel in inches
    plt.subplots(figsize=(h*px, w*px), layout='tight')
    plt.axis('off')
    if c==4:
        plt.imshow( cv2.cvtColor(image, cv2.COLOR_BGRA2RGBA) )
    elif c==1:
        plt.imshow( image, cmap='gray' )
    else:
        plt.imshow( cv2.cvtColor(image, cv2.COLOR_BGR2RGB) )
```

```python
open('mandril.tif', 'wb').write(requests.get("https://sipi.usc.edu/database/download.php?vol=misc&img=4.2.03", allow_redirects=True).content)
mandril = cv2.imread('mandril.tif')
```

```python
h, w, _ = mandril.shape
```

```python
mandrilBGRA = cv2.cvtColor(mandril,cv2.COLOR_BGR2BGRA)
```

```python
origenafin = np.array([[0,0],[w,0],[0,h]])
destinoafin = np.array([[100,50],[400,300],[50,250]])
```

```python
conmarco = mandril.copy()
cv2.polylines(conmarco, [destinoafin], True, (0,0,255), 3) # Marco rojo
cv2.polylines(conmarco, [origenafin], True, (0,255,0),3) # Marco verde
plot(conmarco)
```

```python
A = cv2.getAffineTransform(np.float32(origenafin), np.float32(destinoafin))
afin = cv2.warpAffine(mandrilBGRA.copy(), A, (512,512))
plot(afin)
```

```python
R = cv2.getRotationMatrix2D((256,256), angle=60, scale=0.5)
rot = cv2.warpAffine(mandrilBGRA.copy(), R, (512,512))
plot(rot)
```

```python
origen = np.array([[0,0],[w,0],[w,h],[0,h]])
destino = np.array([[100, 30],[500,100],[400,500],[200,450]])
```

```python
conmarco = mandril.copy()
cv2.polylines(conmarco, [destino], True, (0,0,255), 3) # Marco rojo
cv2.polylines(conmarco, [origen], True, (0,255,0),3) # Marco verde
plot(conmarco)
```

```python
M = cv2.getPerspectiveTransform(np.float32(origen), np.float32(destino))
warp = cv2.warpPerspective(mandrilBGRA.copy(), M, dsize=(512,512))
plot(warp)
```

```python
N = cv2.getPerspectiveTransform(np.float32(destino), np.float32(origen))
praw = cv2.warpPerspective(warp, N, (512,512))
plot(praw)
```
