```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import import_ipynb
import cuia
from deepface import DeepFace
```

```python
oscars = cv2.imread("oscars.jpg")
```

```python
backends = ['opencv', 'ssd', 'mtcnn', 'retinaface']
imagen = oscars.copy()
caras = DeepFace.extract_faces(imagen, detector_backend='retinaface')
for cara in caras:
    f = cara['facial_area']
    c1 = (f['x'], f['y'])
    c2 = (f['x']+f['w'], f['y']+f['h'])
    cv2.rectangle(imagen, c1, c2, (0,255,0),3)
    cv2.putText(imagen, f'{cara["confidence"]*100:.2f}%', (f['x'],f['y']+f['h']), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,0,255), 2, cv2.LINE_AA)
cuia.plot(imagen)
```
