```python
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import requests
```

```python
def plot(image):
    if len(image.shape)==2:
        h, w = image.shape
        c = 1
    else:
        h, w, c = image.shape
    px = 1/plt.rcParams['figure.dpi']  # pixel in inches
    plt.subplots(figsize=(h*px, w*px), layout='tight')
    plt.axis('off')
    if c==4:
        plt.imshow( cv2.cvtColor(image, cv2.COLOR_BGRA2RGBA) )
    elif c==1:
        plt.imshow( image, cmap='gray' )
    else:
        plt.imshow( cv2.cvtColor(image, cv2.COLOR_BGR2RGB) )
```

```python
open('mandril.tif', 'wb').write(requests.get("https://sipi.usc.edu/database/download.php?vol=misc&img=4.2.03", allow_redirects=True).content)
```

```python
mandril = cv2.imread('mandril.tif')
```

[**line**](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2)
> Inicio  (ej. (100,100) )  
> Fin     (ej. (200,300) )  
> Color   (ej. (255,0,0) que representa el color azul en formato BGR)  
> Grosor  (1 por defecto)  
> Tipo    (Tipo de [linea](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#gaf076ef45de481ac96e0ab3dc2c29a777), por defecto cv2.LINE_8)  
> Shift   (Fracci√≥n de pixel de desplazamiento de las coordenadas de los extremos de la linea, por defecto 0)

```python
imagen = mandril.copy()
cv2.line(imagen, (100,100), (200,300), (255,0,0), 4, cv2.LINE_AA)
plot(imagen)
```

[**rectangle**](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#ga07d2f74cadcf8e305e810ce8eed13bc9)
> Esquina1(ej. (250,200) )  
> Esquina2(ej. (450,300) esquina opuesta)  
> Color   (ej. (0,255,0) que representa el color verde en formato BGR)  
> Grosor  (1 por defecto, si el valor es -1 se rellena el rect√°ngulo)  
> Tipo    (Tipo de [linea](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#gaf076ef45de481ac96e0ab3dc2c29a777), por defecto cv2.LINE_8)  
> Shift   (Fracci√≥n de pixel de desplazamiento de las coordenadas de las esquinas, por defecto 0)

```python
imagen = mandril.copy()
cv2.rectangle(imagen, (250,200), (450,300), (0,255,0), -1, cv2.LINE_4, 0)
plot(imagen)
```

[**circle**](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670)
> Centro  (ej. (256,256) )  
> Radio   (ej. 150)  
> Color   (ej. (0,0,255) que representa el color rojo en formato BGR)  
> Grosor  (1 por defecto, si el valor es -1 se rellena el c√≠rculo)  
> Tipo    (Tipo de [linea](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#gaf076ef45de481ac96e0ab3dc2c29a777), por defecto cv2.LINE_8)  
> Shift   (Fracci√≥n de pixel de desplazamiento de las coordenadas del centro y el radio, por defecto 0)

```python
imagen = mandril.copy()
cv2.circle(imagen, (256,256), 150, (0,0,255), 5)
plot(imagen)
```

[**rectangle**](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#ga1ea127ffbbb7e0bfc4fd6fd2eb64263c)
> V√©rtices(Array de v√©rtices de la polil√≠nea, ej [(100,100), (150,250), (300,400)]  
> Cerrado (Valor l√≥gico que indica si debe conectar el primer y √∫ltimo punto)  
> Color   (Color en formato BGR)  
> Grosor  (1 por defecto, si el valor es -1 se rellena el c√≠rculo)  
> Tipo    (Tipo de [linea](https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html#gaf076ef45de481ac96e0ab3dc2c29a777), por defecto cv2.LINE_8)  
> Shift   (Fracci√≥n de pixel de desplazamiento de las coordenadas del centro y el radio, por defecto 0)

```python
imagen = mandril.copy()
cv2.polylines(imagen, [(100,100), (150,250), (300,400)], True, (255,0,255), 5)
plot(imagen)
```
